```python
#Image degraded with noise and restored image

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image
img = cv2.imread('/content/flower.jpg', 0)

# Add Gaussian noise
noise = np.random.normal(0, 25, img.shape)
noisy = img + noise
noisy = np.clip(noisy, 0, 255).astype(np.uint8)

# Restoration
mean_filtered = cv2.blur(noisy, (5,5))
median_filtered = cv2.medianBlur(noisy, 5)

# Display
plt.figure(figsize=(10,6))
plt.subplot(2,2,1), plt.imshow(img, cmap='gray'), plt.title("Original")
plt.subplot(2,2,2), plt.imshow(noisy, cmap='gray'), plt.title("Noisy Image")
plt.subplot(2,2,3), plt.imshow(mean_filtered, cmap='gray'), plt.title("Mean Filter")
plt.subplot(2,2,4), plt.imshow(median_filtered, cmap='gray'), plt.title("Median Filter")
plt.show()
```
