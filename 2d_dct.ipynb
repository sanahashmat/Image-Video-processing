```python
#2d DCT & IDCT image

# Step 1: Import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Step 2: Read image directly from Colab storage
img = cv2.imread('/content/lena.jpg', cv2.IMREAD_GRAYSCALE)

# Step 3: Check if image is loaded
if img is None:
    print("Error: Image not found!")
else:
    # Step 4: Display original image
    plt.figure(figsize=(5,5))
    plt.imshow(img, cmap='gray')
    plt.title("Original Image")
    plt.axis('off')
    plt.show()

    # Step 5: Convert image to float32 (required for DCT)
    img_float = np.float32(img)

    # Step 6: Apply Forward 2D DCT
    dct = cv2.dct(img_float)

    # Step 7: Display DCT magnitude spectrum
    dct_magnitude = np.log(np.abs(dct) + 1)

    plt.figure(figsize=(5,5))
    plt.imshow(dct_magnitude, cmap='gray')
    plt.title("DCT Coefficients (Magnitude)")
    plt.axis('off')
    plt.show()

    # Step 8: Apply Inverse 2D DCT
    img_reconstructed = cv2.idct(dct)

    # Step 9: Display reconstructed image
    plt.figure(figsize=(5,5))
    plt.imshow(img_reconstructed, cmap='gray')
    plt.title("Reconstructed Image (IDCT)")
    plt.axis('off')
    plt.show()
```
