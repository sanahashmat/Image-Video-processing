```python
#2-D DFT & Inverse DFT (Image)

# Step 1: Import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Step 2: Read image directly using file path
# Make sure the image is uploaded to Colab Files panel
img = cv2.imread('/content/lena.jpg', cv2.IMREAD_GRAYSCALE)

# Step 3: Check if image is loaded
if img is None:
    print("Error: Image not found!")
else:
    # Step 4: Display original image
    plt.figure(figsize=(5,5))
    plt.imshow(img, cmap='gray')
    plt.title("Original Image")
    plt.axis('off')
    plt.show()

    # Step 5: Apply Forward 2D DFT
    dft = np.fft.fft2(img)
    dft_shift = np.fft.fftshift(dft)

    # Step 6: Display Magnitude Spectrum
    magnitude_spectrum = 20 * np.log(np.abs(dft_shift) + 1)

    plt.figure(figsize=(5,5))
    plt.imshow(magnitude_spectrum, cmap='gray')
    plt.title("Magnitude Spectrum (DFT)")
    plt.axis('off')
    plt.show()

    # Step 7: Apply Inverse 2D DFT
    inverse_shift = np.fft.ifftshift(dft_shift)
    img_reconstructed = np.fft.ifft2(inverse_shift)
    img_reconstructed = np.abs(img_reconstructed)

    # Step 8: Display reconstructed image
    plt.figure(figsize=(5,5))
    plt.imshow(img_reconstructed, cmap='gray')
    plt.title("Reconstructed Image (IDFT)")
    plt.axis('off')
    plt.show()
```
