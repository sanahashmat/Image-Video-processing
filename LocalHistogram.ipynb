```python
# ==========================================================
# Local Histogram Enhancement using CLAHE
# ==========================================================

# 1. Import required libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt

# ----------------------------------------------------------
# 2. Load Image (Grayscale)
# ----------------------------------------------------------
img = cv2.imread('/content/panda.jpg', cv2.IMREAD_GRAYSCALE)

# If image not found, create a sample image
if img is None:
    img = np.zeros((256,256), dtype=np.uint8)
    cv2.putText(img, "IMAGE", (50,140),
                cv2.FONT_HERSHEY_SIMPLEX,
                2, 180, 4)

plt.figure(figsize=(4,4))
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.show()

# ----------------------------------------------------------
# 3. Global Histogram Equalization (for comparison)
# ----------------------------------------------------------
global_he = cv2.equalizeHist(img)

plt.figure(figsize=(4,4))
plt.imshow(global_he, cmap='gray')
plt.title("Global Histogram Equalization")
plt.axis('off')
plt.show()

# ----------------------------------------------------------
# 4. Local Histogram Enhancement using CLAHE
# ----------------------------------------------------------
# Create CLAHE object
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))

# Apply CLAHE
local_he = clahe.apply(img)

plt.figure(figsize=(4,4))
plt.imshow(local_he, cmap='gray')
plt.title("Local Histogram Enhancement (CLAHE)")
plt.axis('off')
plt.show()

# ----------------------------------------------------------
# 5. Visual Comparison
# ----------------------------------------------------------
plt.figure(figsize=(12,5))

plt.subplot(1,3,1)
plt.imshow(img, cmap='gray')
plt.title("Original")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(global_he, cmap='gray')
plt.title("Global HE")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(local_he, cmap='gray')
plt.title("Local HE (CLAHE)")
plt.axis('off')

plt.show()

# ----------------------------------------------------------
# 6. Histogram Comparison
# ----------------------------------------------------------
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.hist(img.flatten(), 256)
plt.title("Original Histogram")

plt.subplot(1,3,2)
plt.hist(global_he.flatten(), 256)
plt.title("Global HE Histogram")

plt.subplot(1,3,3)
plt.hist(local_he.flatten(), 256)
plt.title("Local HE Histogram")

plt.show()
```
