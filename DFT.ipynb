```python
#Read an image, apply DFT, rotate image

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image (grayscale)
img = cv2.imread('/content/nature.jpg', 0)

# Rotate image by 90 degrees
rotated = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)

# DFT function
def compute_dft(image):
    dft = np.fft.fft2(image)
    dft_shift = np.fft.fftshift(dft)
    magnitude = np.log(1 + np.abs(dft_shift))
    return magnitude

# Compute spectra
spec_original = compute_dft(img)
spec_rotated = compute_dft(rotated)

# Display
plt.figure(figsize=(10,6))
plt.subplot(2,2,1), plt.imshow(img, cmap='gray'), plt.title("Original Image")
plt.subplot(2,2,2), plt.imshow(spec_original, cmap='gray'), plt.title("Spectrum")
plt.subplot(2,2,3), plt.imshow(rotated, cmap='gray'), plt.title("Rotated Image")
plt.subplot(2,2,4), plt.imshow(spec_rotated, cmap='gray'), plt.title("Rotated Spectrum")
plt.show()
```
