```python
# Program 3: 1D and 2D Cosine Signal Generation
# Objective: Generate and plot 1D cosine and 2D cosine signals

import numpy as np
import matplotlib.pyplot as plt

# Define frequency range
w = np.arange(0, np.pi, np.pi/8)

# 1D Cosine Signal
x1 = np.cos(w)

plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.stem(w, x1)
plt.title("1D Cosine Signal")
plt.xlabel("ω")
plt.ylabel("cos(ω)")

# 2D Cosine Signal
W1, W2 = np.meshgrid(w, w)
x2 = np.cos(W1) + np.cos(W2)

plt.subplot(1,2,2)
plt.imshow(x2, cmap='gray')
plt.title("2D Cosine Signal")
plt.colorbar()
plt.show()
```
