```python
#read an image interpolate to its double the size using nearest neighbor interpolation, bilinear & bicubic

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read uploaded image
img = cv2.imread("/content/rose flower.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Original image size
h, w, _ = img.shape

# Resize to double size
nearest = cv2.resize(img, (2*w, 2*h), interpolation=cv2.INTER_NEAREST)
bilinear = cv2.resize(img, (2*w, 2*h), interpolation=cv2.INTER_LINEAR)
bicubic = cv2.resize(img, (2*w, 2*h), interpolation=cv2.INTER_CUBIC)

# Display results
plt.figure(figsize=(12,10))

plt.subplot(2,2,1)
plt.imshow(img)
plt.title("Original Image")
plt.axis("off")

plt.subplot(2,2,2)
plt.imshow(nearest)
plt.title("Nearest Neighbor Interpolation")
plt.axis("off")

plt.subplot(2,2,3)
plt.imshow(bilinear)
plt.title("Bilinear Interpolation")
plt.axis("off")

plt.subplot(2,2,4)
plt.imshow(bicubic)
plt.title("Bicubic Interpolation")
plt.axis("off")

plt.show()
```
