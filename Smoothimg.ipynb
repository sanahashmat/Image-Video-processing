```python
# ==========================================================
# Image Smoothing and Sharpening
# ==========================================================

# 1. Import required libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt

# ----------------------------------------------------------
# 2. Load Image (Grayscale)
# ----------------------------------------------------------
img = cv2.imread('/content/nature.jpg', cv2.IMREAD_GRAYSCALE)

# If image not found, create a sample image
if img is None:
    img = np.zeros((256,256), dtype=np.uint8)
    cv2.putText(img, "IMAGE", (50,140),
                cv2.FONT_HERSHEY_SIMPLEX,
                2, 255, 4)

plt.figure(figsize=(4,4))
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.show()

# ----------------------------------------------------------
# 3. Image Smoothing (Gaussian Low-Pass Filter)
# ----------------------------------------------------------
smooth_img = cv2.GaussianBlur(img, (5,5), 1.0)

plt.figure(figsize=(4,4))
plt.imshow(smooth_img, cmap='gray')
plt.title("Smoothed Image (Gaussian Filter)")
plt.axis('off')
plt.show()

# ----------------------------------------------------------
# 4. Image Sharpening (Laplacian Filter)
# ----------------------------------------------------------
laplacian = cv2.Laplacian(img, cv2.CV_64F)
sharpened_img = img - laplacian
sharpened_img = np.clip(sharpened_img, 0, 255).astype(np.uint8)

plt.figure(figsize=(4,4))
plt.imshow(sharpened_img, cmap='gray')
plt.title("Sharpened Image (Laplacian)")
plt.axis('off')
plt.show()

# ----------------------------------------------------------
# 5. Alternative Sharpening (Unsharp Masking)
# ----------------------------------------------------------
blur = cv2.GaussianBlur(img, (5,5), 1.0)
unsharp = cv2.addWeighted(img, 1.5, blur, -0.5, 0)

plt.figure(figsize=(4,4))
plt.imshow(unsharp, cmap='gray')
plt.title("Sharpened Image (Unsharp Masking)")
plt.axis('off')
plt.show()

# ----------------------------------------------------------
# 6. Visual Comparison
# ----------------------------------------------------------
plt.figure(figsize=(12,6))

plt.subplot(1,4,1)
plt.imshow(img, cmap='gray')
plt.title("Original")
plt.axis('off')

plt.subplot(1,4,2)
plt.imshow(smooth_img, cmap='gray')
plt.title("Smoothed")
plt.axis('off')

plt.subplot(1,4,3)
plt.imshow(sharpened_img, cmap='gray')
plt.title("Laplacian Sharp")
plt.axis('off')

plt.subplot(1,4,4)
plt.imshow(unsharp, cmap='gray')
plt.title("Unsharp Mask")
plt.axis('off')

plt.axis('off')
plt.show()
```
